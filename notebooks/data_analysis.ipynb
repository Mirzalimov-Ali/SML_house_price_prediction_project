```python
import pandas as pd

from matplotlib import pyplot as plt, ticker as mtick
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('../data/engineered/engineered_House_price_dataset.csv')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='sqft_living', y='price', data=df, alpha=0.5)

plt.title("Price vs Living Area")
plt.xlabel("Living Area (sqft)")
plt.ylabel("Price (USD)")

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))

plt.tight_layout()
plt.savefig('../visuals/Price_vs_LivingArea.png')
plt.show()
```

```python
avg_price_by_year = df.groupby('yr_built')['price'].mean().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(x='yr_built', y='price', data=avg_price_by_year)
plt.title("Average Price by Year Built")
plt.xlabel("Year Built")
plt.ylabel("Average Price (USD)")

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.tight_layout()
plt.savefig('../visuals/price_by_yearbuilt.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='grade', y='price', data=df)
plt.title("House Prices by Grade")
plt.xlabel("Grade")
plt.ylabel("Price (USD)")

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.tight_layout()
plt.savefig('../visuals/price_by_grade.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=(df['yr_renovated'] > 0), y='price', data=df)
plt.title("House Prices: Renovated vs Not Renovated")
plt.xlabel("")
plt.ylabel("Price (USD)")
plt.xticks([0,1], ['Not Renovated', 'Renovated'])

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.tight_layout()
plt.savefig('../visuals/renovated_vs_notrenovated.png')
plt.show()
```

```python
avg_price_by_bedrooms = df.groupby('bedrooms')['price'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(x='bedrooms', y='price', data=avg_price_by_bedrooms, palette='viridis')
plt.title("Average Price by Number of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price (USD)")

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.tight_layout()
plt.savefig('../visuals/price_by_bedrooms.png')
plt.show()
```

```python
avg_price_by_floors = df.groupby('floors')['price'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x='floors', y='price', data=avg_price_by_floors, palette='cubehelix')
plt.title("Average Price by Floors")
plt.xlabel("Floors")
plt.ylabel("Average Price (USD)")

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.tight_layout()
plt.savefig('../visuals/price_by_floors.png')
plt.show()
```

```python
avg_price_by_waterfront = df.groupby('waterfront')['price'].mean().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(x='waterfront', y='price', data=avg_price_by_waterfront, palette='Set2')
plt.title("Average Price: Waterfront vs Non-Waterfront")
plt.xlabel("Waterfront")
plt.ylabel("Average Price (USD)")

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.tight_layout()
plt.savefig('../visuals/Price_by_Waterfront.png')
plt.show()
```

```python
fig = px.scatter(df, x='sqft_living', y='price', color='grade', size='sqft_living', hover_data=['bedrooms','bathrooms','floors','waterfront'], title="Interactive Scatter: Price vs Living Area")

fig.update_layout(
    yaxis_title="Price (USD)",
    xaxis_title="Living Area (sqft)",
    yaxis_tickformat='$,.0f'
)

fig.show()
```
